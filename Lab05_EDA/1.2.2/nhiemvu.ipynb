```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load data từ thư mục local (điều chỉnh đường dẫn nếu cần)
houseprices_data = pd.read_csv("../data/HousingPrices-Amsterdam-August-2021.csv")
houseprices_data = houseprices_data[['Zip', 'Price', 'Area', 'Room']]

# Tạo cột mới: Giá trên mỗi m²
houseprices_data['PriceperSqm'] = houseprices_data['Price'] / houseprices_data['Area']

# Kiểm tra data
print(houseprices_data.head())
```

```python
# Sắp xếp theo Price giảm dần
houseprices_sorted = houseprices_data.sort_values('Price', ascending=False)
print(houseprices_sorted.head())
```

```python
# Case 1: Biểu đồ cột giá trung bình theo Zip
plt.figure(figsize=(10, 5))
avg_price_by_zip = houseprices_data.groupby('Zip')['Price'].mean()
avg_price_by_zip.plot(kind='bar')
plt.title('Average Price by Zip Code')
plt.xlabel('Zip Code')
plt.ylabel('Average Price')
plt.xticks(rotation=45)
plt.show()

# Case 2: Biểu đồ cột giá trung bình theo số phòng
plt.figure(figsize=(10, 5))
avg_price_by_room = houseprices_data.groupby('Room')['Price'].mean()
avg_price_by_room.plot(kind='bar')
plt.title('Average Price by Number of Rooms')
plt.xlabel('Number of Rooms')
plt.ylabel('Average Price')
plt.show()

# Case 3: Hai biểu đồ cột song song
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 5))
avg_price_by_zip.plot(kind='bar', ax=ax1)
ax1.set_title('Average Price by Zip Code')
ax1.set_xlabel('Zip Code')
ax1.set_ylabel('Average Price')
ax1.tick_params(axis='x', rotation=45)

avg_price_by_room.plot(kind='bar', ax=ax2)
ax2.set_title('Average Price by Number of Rooms')
ax2.set_xlabel('Number of Rooms')
ax2.set_ylabel('Average Price')
plt.show()
```

```python
# Set style cho Seaborn
sns.set_style("whitegrid")

# Case 1: Boxplot phân bố giá theo Zip
plt.figure(figsize=(10, 5))
sns.boxplot(x='Zip', y='Price', data=houseprices_data)
plt.title('Price Distribution by Zip Code')
plt.xlabel('Zip Code')
plt.ylabel('Price')
plt.xticks(rotation=45)
plt.show()

# Case 2: Boxplot phân bố giá theo số phòng
plt.figure(figsize=(10, 5))
sns.boxplot(x='Room', y='Price', data=houseprices_data)
plt.title('Price Distribution by Number of Rooms')
plt.xlabel('Number of Rooms')
plt.ylabel('Price')
plt.show()

# Case 3: Hai boxplot song song
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 5))
sns.boxplot(x='Zip', y='Price', data=houseprices_data, ax=ax1)
ax1.set_title('Price Distribution by Zip Code')
ax1.set_xlabel('Zip Code')
ax1.set_ylabel('Price')
ax1.tick_params(axis='x', rotation=45)

sns.boxplot(x='Room', y='Price', data=houseprices_data, ax=ax2)
ax2.set_title('Price Distribution by Number of Rooms')
ax2.set_xlabel('Number of Rooms')
ax2.set_ylabel('Price')
plt.show()
```
